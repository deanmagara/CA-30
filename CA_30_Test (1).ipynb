```python
class TrafficLightFSM:
  current_state = None
  A = "A"
  B = "B"
  C = "C"
  D = "D"
  E = "E"

def traffic_light_finite_state_machine(transition_state_tuple_list, advance_signal=False):
  transition_dict = dict(transition_state_tuple_list)

  if TrafficLightFSM.current_state is None:
    if advance_signal:
      TrafficLightFSM.current_state = transition_state_tuple_list[0][0]
    return

  if advance_signal:
    TrafficLightFSM.current_state = transition_dict[TrafficLightFSM.current_state]


transition_state_tuple_list = [
    (TrafficLightFSM.A,    TrafficLightFSM.B),
    (TrafficLightFSM.B,  TrafficLightFSM.C),
    (TrafficLightFSM.C, TrafficLightFSM.D),
    (TrafficLightFSM.D, TrafficLightFSM.E),
    (TrafficLightFSM.E, TrafficLightFSM.A)
]
advance_signal_list = [True, True, True, False, True, False, True, False, False, True]

for advance_signal in advance_signal_list:
  traffic_light_finite_state_machine(
      transition_state_tuple_list=transition_state_tuple_list,
      advance_signal=advance_signal
  )
  print("Traffic Light FSM Current State: {0}\n".format(TrafficLightFSM.current_state))
```
